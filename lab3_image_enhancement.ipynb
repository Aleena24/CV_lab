```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load the image
image = cv2.imread("/content/drive/MyDrive/research_dataset/BSD100/87046.png")
```

```python
# Convert BGR to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

# Negative image

```python
# Create a negative image
negative_image = 255 - image
```

```python
# Plot original and equalized images using matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Equalized image
plt.subplot(1, 2, 2)
plt.imshow(negative_image, cmap='gray')
plt.title('Negative Image')
plt.axis('off')

plt.show()
```

# Gray Level Slicing

```python
# Set the threshold range
low = 100
high = 200
```

```python
# Apply gray level slicing
sliced_image = np.where((image >= low) & (image <= high), 255, 0).astype(np.uint8)
```

```python
# Plot original and equalized images using matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Equalized image
plt.subplot(1, 2, 2)
plt.imshow(sliced_image, cmap='gray')
plt.title('Gray Level Sliced Image')
plt.axis('off')

plt.show()
```

# Histogram Equalization

```python
# Check the type and shape of the image
print("Image dtype:", image.dtype)
print("Image shape:", image.shape)
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Perform histogram equalization
equalized_image = cv2.equalizeHist(gray_image)
```

```python
# Plot original and equalized images using matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Equalized image
plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```
