```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
# Load the continuous image
image = cv2.imread("/content/drive/MyDrive/research_dataset/BSD100/299086.png", cv2.IMREAD_GRAYSCALE)

cv2_imshow(image)
```

# Sampling

Sampling in computer vision refers to the process of converting a continuous image into a discrete grid of pixels. Each pixel represents a sample of the continuous image at a specific location. The sampling rate determines how frequently samples are taken from the continuous image, which in turn affects the resolution of the resulting digital image.

```python
# Perform sampling by selecting pixels at regular intervals
sampled_image = image[::2, ::2]

# Display the sampled image
cv2_imshow(sampled_image)
```

# Quantization

Quantization in computer vision involves reducing the number of distinct intensity levels (or colors) used to represent an image. This process is essential for reducing the amount of data required to store or transmit an image while minimizing the perceptual loss of image quality.



```python
# Perform quantization by reducing the number of intensity levels to 4 bits (16 levels)
quantized_image = np.uint8(image // 16 * 16)

# Display the quantized image
cv2_imshow(quantized_image)
```
